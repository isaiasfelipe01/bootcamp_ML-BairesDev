```python
from PIL import Image
from IPython.display import display

img = Image.open("/content/drive/MyDrive/bootcamp_ml/3 - Algoritmos de Treinamento em ML/senna.jpg")
```

```python
largura, altura = img.size
img_cinza = Image.new("L", (largura, altura))

for i in range(altura):
  for c in range(largura):
    r, g, b = img.getpixel((c, i))
    valor_cinza = int(0.299 * r + 0.587 * g + 0.114 * b) #maneira 1
    cinza = (r + g + b) // 3 #maneira 2
    img_cinza.putpixel((c, i), cinza)
```

```python
img_binaria = Image.new("1", (largura, altura))
for i in range(altura):
  for c in range(largura):
    valor_cinza = img_cinza.getpixel((c, i))
    valor_binario = 1 if valor_cinza > 128 else 0
    img_binaria.putpixel((c, i), valor_binario)
```

```python

display(img)
display(img_cinza)
display(img_binaria)
```
